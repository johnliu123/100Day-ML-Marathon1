## 作業
1. 請將 app_train 中的 CNT_CHILDREN 依照下列規則分為四組，並將其結果在原本的 dataframe 命名為 CNT_CHILDREN_GROUP
    * 0 個小孩
    * 有 1 - 2 個小孩
    * 有 3 - 5 個小孩
    * 有超過 5 個小孩

2. 請根據 CNT_CHILDREN_GROUP 以及 TARGET，列出各組的平均 AMT_INCOME_TOTAL，並繪製 baxplot
3. 請根據 CNT_CHILDREN_GROUP 以及 TARGET，對 AMT_INCOME_TOTAL 計算 [Z 轉換](https://en.wikipedia.org/wiki/Standard_score) 後的分數

```python
a=b=c=d=0
for i,row in app_train.iterrows():
    num=row['CNT_CHILDREN']
    if num==0:
        a+=1
    elif num>0 and num<=2:
        b+=1
    elif num>2 and num<=5:
        c+=1
    else :
        d+=1
print("0個小孩:%d"%a)
print("1-2個小孩:%d"%b)
print("3-5個小孩:%d"%c)
print("5個小孩以上:%d"%d)
```

```python
#1
"""
Your code here
"""
#切分類別資料的範圍會用到離散化pd.cut()
#參考資料
#https://www.jianshu.com/p/123bcf1b9f02

#pandas.cut(x,bins,right=True,labels=None,retbins=False,precision=3,include_lowest=False)

#先設定app_train['CNT_CHILDREN']類別範圍
cut_rule = [-1,0,2,5, app_train['CNT_CHILDREN'].max()]

app_train['CNT_CHILDREN_GROUP'] = pd.cut(app_train['CNT_CHILDREN'].values, cut_rule, include_lowest=True,labels=['0個小孩', '1-2個小孩', '3-5個小孩', '5個小孩以上'])
app_train['CNT_CHILDREN_GROUP'].value_counts()
```

```python
#2-1
"""
Your code here
"""
#請根據 CNT_CHILDREN_GROUP 以及 TARGET，列出各組的平均 AMT_INCOME_TOTAL
#列出(['CNT_CHILDREN_GROUP','TARGET'])欄位資料 並根據['CNT_CHILDREN_GROUP','TARGET']欄位所對應的['AMT_INCOME_TOTAL']資料做平均
app_train.groupby(['CNT_CHILDREN_GROUP','TARGET'])['AMT_INCOME_TOTAL'].mean()
```

```python
grouped_df = app_train.groupby(['CNT_CHILDREN_GROUP','TARGET'])['AMT_INCOME_TOTAL']
grouped_df
```

```python
#2-2
"""
Your code here
"""
plt_column = 'AMT_INCOME_TOTAL'
plt_by =  ['CNT_CHILDREN_GROUP', 'TARGET']

app_train.boxplot(column=plt_column, by = plt_by, showfliers = False, figsize=(12,12))
plt.suptitle('')
plt.show()
```

```python
#3
"""
Your code here
"""
app_train['AMT_INCOME_TOTAL_Z_BY_CHILDREN_GRP-TARGET'] = grouped_df.apply(lambda x:(x-x.mean())/x.std() )

app_train[['AMT_INCOME_TOTAL','AMT_INCOME_TOTAL_Z_BY_CHILDREN_GRP-TARGET']].head()
```
