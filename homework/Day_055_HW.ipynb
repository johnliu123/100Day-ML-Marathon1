### 用 iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('bmh')
import warnings
warnings.simplefilter('ignore')
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
```

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
X.shape,y.shape
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
def plot_cluster(X, y, labels):
    fig = plt.figure(figsize=(8, 6))
    ax = fig.gca(projection='3d')
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.jet)
    ax.set_xlabel(labels[0])
    ax.set_ylabel(labels[1])
    ax.set_zlabel(labels[2])
    ax.view_init(elev=30, azim=225)

# original data labels
#iris.feature_names[1:]取得特徵名稱
plot_cluster(X, y, iris.feature_names[1:])
```

```python
# 2 clusters

estimator = KMeans(n_clusters=2)
estimator.fit(X)
y_pred = estimator.labels_

plot_cluster(X, y_pred, iris.feature_names[1:])
```

```python
# 3 clusters

estimator = KMeans(n_clusters=3)
estimator.fit(X)
y_pred = estimator.labels_

plot_cluster(X, y_pred, iris.feature_names[1:])
```

```python
# 4 clusters

estimator = KMeans(n_clusters=4)
estimator.fit(X)
y_pred = estimator.labels_

plot_cluster(X, y_pred, iris.feature_names[1:])
```
